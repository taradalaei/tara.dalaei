```python
#question 2
!pip install -q pyomo
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.x = Var(domain=NonNegativeIntegers)
model.iset=Set(initialize=[9, 10, 11, 12, 13])
model.y=Var(model.iset, domain=NonNegativeIntegers)

# declare objective
model.obj=Objective(expr=sum(1600*model.y[i] for i in model.iset) + 5000*model.x, sense=minimize)


# declare constraints
def const1_rule(model):
 return model.x + model.y[9] >= 10
model.const1 = Constraint(rule=const1_rule)

def const2_rule(model):
 return model.x + model.y[9] + model.y[10] >= 12
model.const2 = Constraint(rule=const2_rule)

def const3_rule(model):
 return (model.x)/2 + model.y[9] + model.y[10] + model.y[11] >= 14
model.const3 = Constraint(rule=const3_rule)

def const4_rule(model):
 return (model.x)/2 + model.y[9] + model.y[10] + model.y[11] + model.y[12] >= 16
model.const4 = Constraint(rule=const4_rule)

def const5_rule(model):
 return model.x + model.y[10] + model.y[11] + model.y[12] + model.y[13] >= 18
model.const5 = Constraint(rule=const5_rule)

def const6_rule(model):
 return model.x + model.y[11] + model.y[12] + model.y[13] >= 17
model.const6 = Constraint(rule=const6_rule)

def const7_rule(model):
 return model.x + model.y[12] + model.y[13] >= 15
model.const7 = Constraint(rule=const7_rule)

def const8_rule(model):
 return model.x + model.y[13] >= 10
model.const8 = Constraint(rule=const8_rule)

def const9_rule(model):
 return model.x <= 12
model.const9 = Constraint(rule=const9_rule)

def const10_rule(model):
 return 4* sum(model.y[i] for i in model.iset) <= 4*model.x
model.const10 = Constraint(rule=const10_rule)

model.pprint()

!apt-get install -y -qq glpk-utils
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nDecision Variables')
print('x = ', model.x())
print('y = ')
for i in model.iset:
 print(i,':', value(model.y[i]))

model.display()

```

```python
#question 1
!pip install -q pyomo
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.aset = Set(initialize=[1, 2, 3])
model.x = Var(model.aset, domain=NonNegativeIntegers)
model.y1_4 = Var(domain = NonNegativeIntegers)
model.y2_5 = Var(domain = NonNegativeIntegers)
model.y2_6 = Var(domain = NonNegativeIntegers)
model.bset=Set(initialize=[ 5, 6])
model.z=Var(model.bset, domain=NonNegativeIntegers)
model.zprime=Var(model.bset, domain=NonNegativeIntegers)
model.r = Var(domain = NonNegativeIntegers)

# declare objective
expenses = model.r*(6 + 4*4 + 2*4 + 1*2) + model.y1_4*1*1 + model.y2_5*0.8*5 + model.y2_6*0.3*5 + model.z[5]*4 + model.z[6]*3
profits = (model.x[1] - model.y1_4)*7 + (model.x[2] - model.y2_5 - model.y2_6)*6 + model.x[3]*4 + model.y1_4*3 + (model.y2_5 - model.z[5])*20 + (model.y2_6 - model.z[6])*35
model.obj=Objective(expr= profits - expenses , sense=maximize)


# declare constraints
def const1_rule(model):
 return model.x[1] - model.y1_4 <= 1200
model.const1 = Constraint(rule=const1_rule)
def const2_rule(model):
 return (model.x[2] - model.y2_5 - model.y2_6) <= 300
model.const2 = Constraint(rule=const2_rule)
def const3_rule(model):
 return model.y2_5 - 1000 == model.z[5] - model.zprime[5]
model.const3 = Constraint(rule=const3_rule)
def const4_rule(model):
 return model.y2_6 - 800 == model.z[6] - model.zprime[6]
model.const4 = Constraint(rule=const4_rule)
def const5_rule(model):
 return model.r <= 3000
model.const5 = Constraint(rule=const5_rule)
def const6_rule(model):
 return model.x[1] == 4*model.r
model.const6 = Constraint(rule=const6_rule)
def const7_rule(model):
 return model.x[2] == 2*model.r
model.const7 = Constraint(rule=const7_rule)
def const8_rule(model):
 return model.x[3] == model.r
model.const8 = Constraint(rule=const8_rule)
def const9_rule(model):
 return model.x[1] - model.y1_4 >= 0
model.const9 = Constraint(rule=const9_rule)
def const10_rule(model):
 return (model.x[2] - model.y2_5 - model.y2_6) >= 0
model.const10 = Constraint(rule=const10_rule)


model.pprint()

!apt-get install -y -qq glpk-utils
#SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()
result= SolverFactory('glpk').solve(model, 'glpk', keepfiles=True, tee=True)
# display solution
print('---------------------------------------------------------------------')

print('\nDecision Variables')
print('raw material = ', model.r())

print('number of manufactured products  = ')
for i in model.aset:
 print(i,':', value(model.x[i]))

print('number of product 1 proccessed into product 4 = ', model.y1_4())
print('number of product 2 proccessed into product 5 = ', model.y2_5())
print('number of product 2 proccessed into product 6 = ', model.y2_6())

print('number of destroyed products = ')
for i in model.bset:
 print(i,':', value(model.z[i]))

print('maximum z = ')
print(model.obj())
print('---------------------------------------------------------------------')
model.display()
```
